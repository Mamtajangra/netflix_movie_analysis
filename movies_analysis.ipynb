## categorize the movies feedback in popular,average,non popular,below average

```python
def categorize(df,col,labels):
    edges = [df[col].describe()["min"],                
            df[col].describe()["25%"],
            df[col].describe()["50%"],
            df[col].describe()["75%"],
            df[col].describe()["max"]]
    df[col]= pd.cut(df[col],edges,labels = labels, duplicates = "drop")      ## cut function is used to categorize it
    return df    
```

```python
labels = ["non_popular","below_average","average","popular"]      ## labelling it 
categorize(df,"Vote_Average",labels)
df["Vote_Average"].unique()
```
